<h3>2.4.2 Imputation of Missing Data</h3>

<p>Using the identified strategy listed in the previous section, I will be using the identified strategy on the data column with missing data.</p>

```python
# ----------- String Filling ----------- #
merged_df['company_name'] = merged_df['company_name'].fillna('Unknown')

# ----------- Integer Filling ----------- #
merged_df['views'] = merged_df['views'].fillna(0)
merged_df['applies'] = merged_df['applies'].fillna(0)
merged_df['remote_allowed'] = merged_df['remote_allowed'].fillna(0)

# ----------- Mode Imputation ----------- #
merged_df['pay_period'] = merged_df['pay_period'].fillna(merged_df['pay_period'].mode()[0])
merged_df['formatted_experience_level'] = merged_df['formatted_experience_level'].fillna(
    merged_df['formatted_experience_level'].mode()[0]
)

# ----------- Function for Interpolation by Groups ----------- #
def interpolate_salary_by_group(df, column):
    # ----------- Sorting of Data ----------- #
    df = df.sort_values(by = ['pay_period', 'formatted_experience_level', 'listed_time'])
    
    # ----------- Interpolation Process ----------- #
    df[column] = df.groupby(['pay_period', 'formatted_experience_level'])[column] \
                   .transform(lambda x: x.interpolate(method = 'linear', limit_direction = 'both'))
    
    # ----------- Returning of Final Results ----------- #
    return df

# ----------- Interpolation for max_salary and min_salary ----------- #
merged_df = interpolate_salary_by_group(merged_df, 'max_salary')
merged_df = interpolate_salary_by_group(merged_df, 'min_salary')

# ----------- Placing Missing Values of the Merged Dataframe Into A Dataframe ----------- #
mergedDfMissingData = pd.DataFrame({'Missing Values': merged_df.isnull().sum().values}, index = merged_df.columns)

# ----------- Function for Colour Coding ----------- #
def colour_code(val):
    if val > 0:
        return 'background-color: red; color: white'
    else:
        return 'background-color: green; color: white'

# ----------- Calling Function for Colour Coding ----------- #
mergedDfMissingData = mergedDfMissingData.style.map(colour_code, subset = ['Missing Values'])

# ----------- Displaying Missing Values----------- #
mergedDfMissingData
```

<h3>2.9.1 Handling Outliers Using IQR Method</h3>

<p>The IQR Method requires the establishment of a Upper Fence and Lower Fence such that any data which are above the Upper Fence will be deemed invalid and illegitimate and will be removed whereas any data which are below the Lower Fence will be deemed likewise and removed. In the cells below, I will be removing Outliers using the IQR Method and subsequently comparing the effectiveness and results with the Z-Score Method. The formula for establishing the Upper Fence, Lower Fence and IQR are listed below.</p>

$$\text{IQR} = \text{Q3} - \text{Q1}$$
$$\text{Lower Fence} = \text{Q1} - 1.5 \times \text{IQR}$$
$$\text{Upper Fence} = \text{Q3} + 1.5 \times \text{IQR}$$
<br/>
<p>With the formulas listed, I will proceed with identifying and removing the Outliers.</p>

<p>In the previous section, I have used the IQR method to identify and address Outliers. After removing Outliers using the Z-score Method, I will then cross refer the result and verify which method is most suitable. The formula for Z-Score is indicated below.</p>

$$
Z = \frac{X - \mu}{\sigma}
$$

\begin{align*}
Z & : \text{Z-Score (standardised value)} \\
X & : \text{Observed data point} \\
\mu & : \text{Mean of the data} \\
\sigma & : \text{Standard deviation of the data}
\end{align*}

<br>
<p>With the formula indicated, I will proceed to identify the Outliers and remove the Outliers in the following cells.</p>

<h3>3.2.1 Relationship Between Apply to View Ratio and Salary Range (Boxplot)</h3>

<p>I will be using a boxplot to see the relationship between Apply to View Ratio and Salary Range to see whether there is higher application and views for job listing with higher salary. I will also be evaluating their relationship after the plotting. The summarised details are indicated below.</p>
<br/>
<p>Chart Type: <strong><u>Boxplot</u></strong></p>
<p>X-axis Variable: <strong><u>salary_range (Will be grouped)</u></strong></p>
<p>Y-axis Variable: <strong><u>apply_view_ratio</u></strong></p>
<p>Objective: <strong><u>To view application rates for each salary range.</u></strong></p>

```python
# ----------- Filtering 0 Values ----------- #
filtered_df = merged_df[merged_df['apply_view_ratio'] > 0].copy()

# ----------- Binning Ranges ----------- #
bin_edges = [0, 1000, 3000, float('inf')]
salary_bins = ['Low (0-1000)', 'Medium (1000-3000)', 'High (3000+)']

# ----------- Assigning of Bins ----------- #
filtered_df['salary_range_group'] = pd.cut(filtered_df['salary_range'], bins = bin_edges, labels = salary_bins)

# ----------- Data Preparation for Plotting ----------- #
boxplot_data = [filtered_df.loc[filtered_df['salary_range_group'] == group, 'apply_view_ratio'].dropna()
                for group in salary_bins]

# ----------- Plotting of Boxplot ----------- #
fig, ax = plt.subplots(figsize=(10, 7))

box = ax.boxplot(
    boxplot_data,
    labels = salary_bins,
    patch_artist = True,
    boxprops = dict(facecolor = 'white', color = 'black', linewidth = 1.5),
    medianprops = dict(color = 'red', linewidth = 2),
    whiskerprops = dict(color = 'black', linestyle = '--'),
    capprops = dict(color = 'black'),
    showfliers = False
)

# ----------- Colour Customisation ----------- #
box_colors = ['coral', 'lightblue', 'lightgreen']

for patch, color in zip(box['boxes'], box_colors):
    patch.set_facecolor(color)

# ----------- Mean Values Annotations ----------- #
for i, group in enumerate(boxplot_data):
    mean_value = group.mean()
    plt.scatter(i + 1, mean_value, color = 'darkblue', marker = 'D', s = 40, label = 'Mean' if i == 0 else "")
    plt.text(i + 1, mean_value, f'{mean_value:.2f}', ha = 'center', va = 'bottom', fontsize = 10, color = 'darkblue')

# ----------- Grid and Aesthetic Improvements ----------- #
plt.title("Boxplot of Apply-to-View Ratio Across Salary Ranges", fontsize = 16, weight = 'bold', pad = 15)
plt.xlabel("Salary Range Group", fontsize = 12, labelpad = 10)
plt.ylabel("Apply-to-View Ratio", fontsize = 12, labelpad = 10)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.legend(loc = 'upper right', fontsize = 10, frameon = False)

# ----------- Displaying the Final Plot ----------- #
plt.tight_layout()
plt.show()
```

<p>With reference to the boxplot above, I am able to see that Job Postings with high Salary Range (\$3000+) and low Salary Range (\$0 to \$1000) generally have a higher Apply-to-View Ratio. <h3>3.2.5 Apply-to-View Ratio Over Time (Line Chart)</h3>

<p>To visualise how the Apply-to-View Ratio changes over time, I will be using a Line Chart to see if putting a Job Posting up does attract more applicants. I will then subsequently indicate the implications for the two main stakeholders after the visualisation. The summarised details are listed below.</p>
<br/>
<p>Chart Type: <strong><u>Line Chart</u></strong></p>
<p>X-axis Variable: <strong><u>postingDuration</u></strong></p>
<p>Y-axis Variable: <strong><u>apply_view_ratio</u></strong></p>
<p>Objective: <strong><u>To see if longer posing duration affects the apply-to-view ratio.</u></strong></p>
<br/>
<p>With the information listed, I will proceed to plot the Line Chart in the cell below.</p>

```python
# ----------- Calculate Mean and Grouping of Data ----------- #
duration_apply_ratio = merged_df.groupby('postingDuration')['apply_view_ratio'].mean()

# ----------- Rolling Average for Smooth Line ----------- #
rolling_window = 7
smoothed_ratio = duration_apply_ratio.rolling(window = rolling_window, min_periods=1).mean()

# ----------- Key Points to Highlight ----------- #
max_point = duration_apply_ratio.idxmax(), duration_apply_ratio.max()
min_point = duration_apply_ratio.idxmin(), duration_apply_ratio.min()

# ----------- Plotting the Line Chart ----------- #
plt.figure(figsize = (12, 7))
plt.plot(duration_apply_ratio.index, duration_apply_ratio.values, marker = 'o', linestyle = '--', color = 'purple', alpha = 0.5, label = "Original Data")
plt.plot(duration_apply_ratio.index, smoothed_ratio, color='blue', linewidth=2, label="Smoothed Line (Rolling Avg)")

# ----------- Highlight Key Points ----------- #
plt.scatter(*max_point, color = 'red', s = 80, label = f"Peak (Day {max_point[0]})")
plt.scatter(*min_point, color = 'green', s = 80, label = f"Lowest (Day {min_point[0]})")

# ----------- Annotating the Key Points ----------- #
plt.text(max_point[0], max_point[1] + 0.005, f"Max: {max_point[1]:.2f}", fontsize = 10, color = 'red', weight = 'bold', ha = 'right')
plt.text(min_point[0], min_point[1] - 0.01, f"Min: {min_point[1]:.2f}", fontsize = 10, color = 'green', weight = 'bold', ha = 'right')

# ----------- Adding Title, Labels, and Grid ----------- #
plt.title("Posting Duration vs Apply-to-View Ratio", fontsize = 16, weight = 'bold')
plt.xlabel("Posting Duration (Days)", fontsize = 12)
plt.ylabel("Average Apply-to-View Ratio", fontsize = 12)
plt.axhline(y = smoothed_ratio.mean(), color = 'orange', linestyle = '--', linewidth = 1.5, label = "Overall Average")
plt.legend()
plt.grid(True, linestyle = '--', alpha = 0.5)

# ----------- Display the Plot ----------- #
plt.tight_layout()
plt.show()
```

<p>With reference to the output of the line graph above, I am able to see that the Posting Duration with the highest Apply-to-View Ratio is 120 Days (4 Months). I am also able to see that generally the Apply-to-View Ratio generally increases as the Posting Duration increases. This could be due to longer duration allowing more Job-seekers to find the Job Post and ultimately lead to higher application rates. I will be listing the implications for both the two main stakeholders below.</p>
<p><strong><u>Employer</u></strong></p>
<ul>
    <li>Put Job Posting for up to 120 Days (4 Months) if large amount of applicants is required.</li>
    <li>Put Job Posting for longer than 25 Days as before 25 days, most apply-to-view ratio is relaively low.</li>
</ul>
<p><strong><u>Job-seeker</u></strong></p>
<ul>
    <li>Job-seekers should try to apply for Job Postings before 120 Days as that is when most applications come in and potentially push your application behind, making the application stand out less.</li>
    <li>Job-seekers should try to apply for Job Postings when application is at the lowest (0) so as to stand out and potentially expedite the Interview and Evaluation process by Employers.</li>
</ul>
